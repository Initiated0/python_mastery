```python
# making custom containers

class TagCloud:
  def __init__(self):
    self.tags = {}


  def add(self, tag):
    self.tags[tag.lower()] = self.tags.get(tag.lower(), 0) + 1

  def __getitem__(self, tag):
    return self.tags.get(tag.lower(), 0)

  def __setitem__(self, tag, count):
    self.tags[tag.lower()] = count
  
  def __len__(self):
    return len(self.tags)

  def __iter__(self):
    return iter(self.tags)

cloud = TagCloud()
cloud.add("Python")
cloud.add("python")
cloud.add("python")
print(cloud.tags)

cloud["PYTHON"] = 10

print(cloud.tags)


# you need to play with this shit, otherwise you won't get the full picture

print(cloud["Python"])
print(cloud.tags["python"])
#print(cloud.tags["Python"]) # gives key error because everything is stored as lower case letters
```

```python
class InvalidOpError(Exception):
  pass

class Stream(object):
  def __init__(self):
    self.opened = False

  def open(self):
    if self.opened:
      raise InvalidOpError("Stream's already open!!!")
    self.opened = True

  def close(self):
    if not self.opened:
      raise InvalidOpError("Stream's already closed!!!")
    self.opened = False
    

  class FileStream(Stream):
    def read(self):
      print("reading data from a file")

  class NetworkStream(Stream):
    def read(self):
      print("reading data from a network")

            
```

```python
# Turning the previous example into abstrct method

from abc import ABC, abstractmethod # abc -> abstract base class

class InvalidOpError(Exception):
  pass

class Stream(ABC):
  def __init__(self):
    self.opened = False

  def open(self):
    if self.opened:
      raise InvalidOpError("Stream's already open!!!")
    self.opened = True

  def close(self):
    if not self.opened:
      raise InvalidOpError("Stream's already closed!!!")
    self.opened = False
  
  @abstractmethod
  def read(self):
    pass

  class FileStream(Stream):
    def read(self):
      print("reading data from a file")

  class NetworkStream(Stream):
    def read(self):
      print("reading data from a network")


  class MemoryStream(Stream):
    # atleast need to implement read() to make it concrete rather than abstract
    def read(self):
      print("reading data from memory stream")

  # stream = Stream() # this gives -> TypeError: Can't instantiate abstract class Stream with abstract methods read
  # stream.open()


  stream = MemoryStream() # this gives -> TypeError: Can't instantiate abstract class MemoryStream with abstract methods read
  stream.open()
  stream.read()


            
```

```python
from types import MethodDescriptorType
# polymorphism

from abc import ABC, abstractmethod

class UIControl(ABC):
  @abstractmethod
  def draw(self):
    pass


class TextBox(UIControl):
  def draw(self):
    print("textbox")


class DropDownList(UIControl):
  def draw(self):
    print("dropdownlist")

# this is the polymorphism power
# this draw function that uses all the other classes simultaneously is amazing

def draw(controls):
  for control in controls:
    control.draw()


ddl = DropDownList()
tb = TextBox()
isinstance(ddl, UIControl)
# draw(ddl)
# draw(tb)

draw([ddl, tb])


# this is a classic example of POLYMORPHISM
# how do you do it

# 1. Create an abstract class using ABC
# 2. Create a method with abstractmethod
# 3. All children of that abstract methde will have a implimentation of the abstract Method 
# 4. Create a single method of the same abstract method which will look at the instance it receives and perform accordingly as its class
```
